```python
#using jupyter notebook
import pandas as pd
from selenium import webdriver
import time
import warnings
from bs4 import BeautifulSoup

warnings.filterwarnings(action='ignore')
```

```python
newsUrl = 'https://sports.news.naver.com/wfootball/record/index?category=epl&year=2021&tab=team'
wd = webdriver.Chrome('chromedriver.exe')
wd.get(newsUrl)
time.sleep(1)
```

```python
page = wd.page_source
```

```python
team_rank_list = BeautifulSoup(page, "html.parser")
team_rank_list = team_rank_list.select('#wfootballTeamRecordBody>table>tbody>tr')
team_rank_list
```

```python
num = []
name = []
for team in team_rank_list:
    num.append(team.select('td.num > div.inner > strong')[0].text)
    name.append(team.select('.name')[0].text)
    
```

```python
num
```

```python
name
```

```python
rank_series = pd.Series(num)
team_series = pd.Series(name)
```

```python
df = pd.DataFrame({'순위': rank_series, '팀':team_series})
```
