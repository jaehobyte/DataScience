```python
import pandas as pd
import folium
```

```python
filepath = '/content/drive/MyDrive/DataScience/CoffeeBean.xlsx'
CB = pd.read_excel(filepath)
CB
```

```python
CB.drop('Unnamed: 0',axis=1,inplace = True)
CB
```

```python
addr = []
for address in CB['address']:
  addr.append(address.split())
```

```python
addr
```

```python
addr2 = []
for i in range(len(addr)):
  if addr[i][0] == '서울':
    addr[i][0] = '서울특별시'
  elif addr[i][0] == '서울시':
    addr[i][0] = '서울특별시'
  elif addr[i][0] == '경기':
    addr[i][0] = '경기도'
  elif addr[i][0] == '경남':
    addr[i][0] = '경상남도'
  elif addr[i][0] == '광주':
    addr[i][0] = '광주광역시'
  elif addr[i][0] == '부산':
    addr[i][0] = '부산광역시'
  elif addr[i][0] == '부산시':
    addr[i][0] == '부산광역시'
  elif addr[i][0] == '인천':
    addr[i][0] = '인천광역시'
  elif addr[i][0] == '인천시':
    addr[i][0] = '인천광역시'
  elif addr[i][0] == '제주도':
    addr[i][0] = '제주특별자치도'
  elif addr[i][0] == '충남':
    addr[i][0] = '충청남도'
  elif addr[i][0] == '충북':
    addr[i][0] = '충청북도'

  addr2.append(' '.join(addr[i]))
```

```python
addr2
```

```python
CB2 = CB.copy() #깊은 복사 (얕은 복사는 원본 data에 손상을 줌)
```

```python
CB2['address2'] = addr2
```

```python
CB2.drop('address',axis=1, inplace=True)
```

```python
CB2.rename(columns={'address2':'address'}, inplace = True)
```
